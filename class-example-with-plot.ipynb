```python
def f(a):
    return a**3-2*a-5
```

```python
def df(a):
    return 3*a**2-2
```

```python
x0=0
tolerance=1e-6
max_iterations=100
```

```python
for i in range(max_iterations):
    fa=f(x0)
    dfa=df(x0)
    x1=x0-fa/dfa
    if abs(f(x1))<tolerance:
        print(f'root at x:{x1:.6f}')
        break
    else:
        x0=x1
```
